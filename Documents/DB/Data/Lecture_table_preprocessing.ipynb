# Lecture table preprocessing
- columns: id, subject, professor, time, weekday, room_id

```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
# 원본 데이터 불러오기
room = pd.read_csv('room.csv', encoding='cp949')
room.head()
```

```python
# 원본 데이터 불러오기
classes = pd.read_excel('종합강의시간표목록.xlsx')
classes.head()
```

```python
# 필요한 열만 추출하기
lecture = classes[['교과목명', '교원명', '요일/시간', '강의실']]
lecture.info()
```

```python
# 데이터 내용 살펴보기
lecture.head()
```

```python
# 데이터 열 이름 바꾸기
lecture.columns = ['subject', 'professor', 'time', 'room']
lecture.head()
```

time 열의 경우 시간표의 수업 시간을 의미한다. 이 중 사이버강의의 경우 수업 시간이 정해지지 않아서 time 값이 NaN이다. 따라서 해당 값을 제거한다

```python
# time 열의 결측치 제거하기
df = lecture.dropna(axis=0)
df.info()
```

```python
# 데이터 복제하기
df = df[['subject', 'professor']]
time = lecture['time']
room = lecture['room']
```

```python
len('월4.0-5.0/12:00-13:30')
len('수3.0-4.0/11:00-12:30')
```

```python
time = time.str.split(',')
time
```

```python
time = time.apply(lambda x: pd.Series(x))
time
```

```python
time.stack()
```

```python
time.stack().reset_index(level=1, drop=True)
```

```python
result = time.stack().reset_index(level=1, drop=True).to_frame('time')
result
```

```python
result = df.merge(result, left_index=True, right_index=True, how='left')
result
```

주2회 수업의 경우 강의실이 2곳인 경우도 있기 때문에 강의실을 분할한다.

```python
room = room.str.split(',')
room
```

```python
room = room.apply(lambda x:pd.Series(x))
room
```

```python
room = room.stack().reset_index(level=1, drop=True).to_frame('room')
room
```

```python
result = result.merge(room, left_index=True, right_index=True, how='left')
result
```
