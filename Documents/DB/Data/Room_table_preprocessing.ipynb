## Room table

```python
# 필요 라이브러리 import
import pandas as pd
import numpy as np
import openpyxl
```

```python
room = pd.read_excel('교내시설안내.xlsx')
room.head()
```

```python
room.info()
```

```python
# 시설명에 널값 13개 존재
room['시설명'].isnull().sum()
```

```python
room[room['시설명'].isnull()]
```

```python
# 시설명에 결측치 있는 행 삭제
room.dropna(subset=['시설명'], inplace=True)
room.info()
```

```python
del room['시설분류']
del room['비고']
```

```python
room.columns = ['id', 'building_name', 'room_name', 'capacity', 'equip_info', 'facility_info']
```

```python
room
```

```python
room['floor'] = room['room_name'].str[0]
room
```

```python
room['floor'].unique()
```

```python
room.to_csv('room.csv')
```
